```python
import pandas as pd
pop=pd.read_csv('popmonth.csv', encoding='cp949', index_col=0, thousands=',')
pop
```

```python
pop.columns =  ['총인구수', '세대수', '세대당인구','남자인구수', '여자인구수','남여비율']
pop.head()
```

```python
pop.index[1]
```

```python
pop.index = [pop.index[i].split()[0] for i in range(len(pop))]
pop
```

```python
import matplotlib.pyplot as plt #이거1
import seaborn as sns   #이거1
sns.scatterplot(pop['총인구수']) #샵 주석
plt.xticks(rotation=45) #이걸 없애면 글자가 겹쳐짐.
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
pop1 = pop.sort_values('총인구수')
sns.barplot(x=pop1.index, y='총인구수',data =pop1)
plt.xticks(rotation=90)
plt.show
```
